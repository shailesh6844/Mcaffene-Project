# Q.write a python program to find maximum occuring character in a string input= abracadabra

```python
a="abracadabra"
lst=[]
for i in a:
    lst.append(a.count(i))
lst
pairs=list(zip(lst,a))
print(pairs)
print(max(lst))
print("maximum occuring character in given string : " ,max(pairs))
```

```python
from collections import Counter
a= "abracadabra"
print(a)

result= Counter(a)
result= max(result, key=result.get)

print("maximum occuring character in given string : ",result)
```

# Q.write a function to find maximum element in list

```python
def max_no(list1):
    max_no = list1[0]  
    for i in list1:
        if i > max_no:
            max_no = i  

    return max_no


a = [25, 89, 35, 84, 14, 90]
res= max_no(a)
print(f"Maximum element of given list is:  {res}")
```

# Q.What is the quantum month on month from July to November?


```python
# Filter the dataset to include only complaints from July to November
months = ['July', 'August', 'September', 'October', 'November']
filtered_complaints = data[data['Month'].isin(months)]

# Count the number of complaints for each month
complaints_per_month = filtered_complaints['Month'].value_counts().reindex(months)

# Plotting
plt.figure(figsize=(10, 6))
complaints_per_month.plot(kind='bar', color='purple')
plt.title('Quantum of Complaints Month on Month (July to November)')
plt.xlabel('Month')
plt.ylabel('Number of Complaints')


plt.show()
```

# Q.What is the quantum of delivery complaints (MOM) from July to November?


```python
# Filter out rows where the complaint type is 'Delivery' and month is from July to November
delivery_complaints_mom = data[
    (data['Query Class'] == 'Delivery') &
    (data['Month'].isin(['July', 'August', 'September', 'October', 'November']))
]

# Count the number of delivery complaints per month
complaints_count = delivery_complaints_mom.groupby('Month').size()

# Plotting
plt.figure(figsize=(10, 6))
complaints_count.plot(kind='barh', color='green')
plt.title('Delivery Complaints (MOM) from July to November')
plt.xlabel('Month')
plt.ylabel('Count')


plt.show()
```

# Q.Which product has highest complaint from Batch - B002?


```python
# Filter the dataset to include only complaints related to Batch B002
batch_b002_complaints = data[data['Batch Number/Fcailities'] == 'B002']

# Count the number of complaints for each product within Batch B002
complaints_per_product = batch_b002_complaints['Product Name'].value_counts()

# Identify the product with the highest number of complaints
highest_complaint_product = complaints_per_product.idxmax()

# Plotting
plt.figure(figsize=(10, 6))
complaints_per_product.plot(kind='bar', color='blue')
plt.title('Complaints for Products in Batch B002')
plt.xlabel('Product Name')
plt.ylabel('Number of Complaints')


plt.show()

print(f"The product with the highest number of complaints in Batch B002 is: {highest_complaint_product}")
```
